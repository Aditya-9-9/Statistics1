1.What is a random variable in probability theory?
 - Random variables can represent outcomes numerically.
 - They allow us to define probability distributions (like PMF for discrete and PDF for continuous RVs).
 - They are essential for calculating expected value, variance, and other statistical measures.

2.What are the types of random variables?
 - Discrete Random Variables
 - Continuous Random Variables

3.What is the difference between discrete and continuous distributions?
 - Discrete Distributions
  - Describe the probabilities of a discrete random variable.
  - The variable can take on finite or countably infinite values.
  - Probabilities are assigned to specific outcomes using a Probability Mass Function (PMF).
 - Continuous Distributions
   - Describe the probabilities of a continuous random variable.
   - The variable can take any value in an interval (uncountably infinite values).
   - Probabilities are described using a Probability Density Function (PDF).

4.What are probability distribution functions (PDF)?
 - A Probability Distribution Function (PDF) describes how probabilities are distributed for a random variable.
 - PDF is not the probability directly, but a function whose area gives probability.
 - Graphically, the PDF is a smooth curve (e.g., bell-shaped for normal distribution).

5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
 - Probability Density Function (PDF)
  - Describes the relative likelihood that the random variable takes on a particular value.
  - It is not the probability directly, but the area under the curve gives probability.
 - Cumulative Distribution Function (CDF)
  - Describes the cumulative probability up to a value x.
  - S-shaped curve that starts at 0 and approaches 1.

6.What is a discrete uniform distribution?
 - A Discrete Uniform Distribution is a type of probability distribution where all possible outcomes are equally likely.

7.What are the key properties of a Bernoulli distribution/
 - Discrete random variable with only 2 outcomes: 0 and 1.
 - The Bernoulli distribution is the foundation for other distributions like the Binomial distribution (sum of many Bernoulli trials).

8.What is the binomial distribution, and how is it used in probability?
 - The Binomial Distribution is one of the most important discrete probability distributions in statistics.
 - Defective items in a batch (quality control).
 - Number of customers who buy a product.
 - Number of correct answers in a multiple-choice test (guessing).

9.What is the Poisson distribution and where is it applied?
 - The Poisson distribution is a discrete probability distribution used to model the number of events occurring in a fixed interval of time, distance, area, or any other measurement, when these events occur randomly and independently at a constant average rate.
 - Number of phone calls at a call center per hour.
 - Number of accidents at a traffic junction per day.

10.What is a continuous uniform distribution?
 - The Continuous Uniform Distribution is a type of probability distribution where all values in a given interval are equally likely. It is the continuous counterpart of the discrete uniform distribution.

11.What are the characteristics of a normal distribution?
 - Bell-shaped and Symmetric
 - Defined by Two Parameters
 - Probability Density Function (PDF)

12.What is the standard normal distribution, and why is it important?
 - The Standard Normal Distribution is a special case of the normal distribution that is centered at zero and has a standard deviation of one.
 - Simplifies Probability Calculations
 - Foundation for Statistical Methods
 - Symmetry and Known Probabilities

13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
 - The Central Limit Theorem (CLT) is one of the most fundamental concepts in statistics because it explains why the normal distribution is so widely used.
 - Justifies the Use of Normal Distribution
 - Foundation for Inference

14.How does the Central Limit Theorem relate to the normal distribution?
 - The CLT states that the sampling distribution of the sample mean becomes approximately normal as the sample size
n increases, regardless of the population’s original shape (skewed, uniform, etc.), as long as the population has a finite mean and variance.
 - This means that even if the population is not normally distributed, the distribution of the sample means will be normal for large
n.

15.What is the application of Z statistics in hypothesis testing?
 - The Z-statistic (or Z-test) is widely used in hypothesis testing when we want to test claims about a population mean or proportion. It is based on the standard normal distribution and is applicable when certain conditions are met.
 - Quality control (checking if the mean of a batch matches the standard).
 - Comparing sample proportions or means to a known standard.

16.How do you calculate a Z-score, and what does it represent?
 - A Z-score is a standardized score that tells you how many standard deviations a data point or sample statistic is away from the mean.
 - Positive Z-score (+): Value is above the mean.
 - Negative Z-score (−): Value is below the mean.
 - Z = 0: Value is exactly at the mean.

17.What are point estimates and interval estimates in statistics?
 - Point Estimates
   - A point estimate is a single value calculated from a sample that is used to estimate an unknown population parameter.
 - Interval Estimates
   - An interval estimate gives a range of plausible values for the population parameter. It is usually expressed as a confidence interval (CI).

18.What is the significance of confidence intervals in statistical analysis/
 - Confidence intervals (CIs) are crucial in statistical analysis because they provide more information than just a single point estimate
 - They Show the Range of Plausible Values
 - They Quantify Uncertainty
 - They Indicate Precision

19.What is the relationship between a Z-score and a confidence interval?
 - The Z-score and confidence interval (CI) are closely related because the Z-score determines the width (margin of error) of a confidence interval when using the standard normal distribution.

20.How are Z-scores used to compare different distributions?
 - Different distributions may be on different scales (e.g., test scores out of 100 vs. 200).
 - A Z-score standardizes the data, converting it into a common scale measured in standard deviations from the mean.
 - This allows direct comparison across distributions.

21.What are the assumptions for applying the Central Limit Theorem/
 - The Central Limit Theorem (CLT) is powerful, but it relies on some key assumptions to hold true. These assumptions ensure that the sampling distribution of the sample mean will approximate a normal distribution.
 - Random Sampling
 - Independence

22.What is the concept of expected value in a probability distribution?
 - The expected value (also called the mean or expectation) of a probability distribution is the long-term average value you would expect if you repeated a random experiment many times. It represents the "center" or the average outcome of the distribution.

23.How does a probability distribution relate to the expected outcome of a random variable?
 - The probability distribution defines the likelihood of each outcome.
 - The expected value is the average of those outcomes, weighted by their probabilities.
 - Without the probability distribution, you can’t calculate the expected outcome.
